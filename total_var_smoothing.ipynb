# Total Variation Smoothing

$\textbf{Problem statement:}$ A good way to denoise signals is total variation smoothing. In 1D we are given an input vector $\textbf{y} \in \mathbb{R}^N$ and want to find $$argmin_{\mu} || \textbf{y}-\mu||_2 ^2 + \lambda \sum_{i=1}^{N-1} |\mu_i - \mu_{i+1}|$$

Here we run this on data of the form $y_n = \mu_n + \epsilon_n$ with iid Gaussian $\epsilon_n$ and a piecewise constant function given by the values $\mu_n$.

The goal here is to implement an algorithm finding a solution and visualize examples of results. Also, we want to compare this to other smoothing algorithms. 

# 1D problem:

### 1. Problem generation
In this section,  we create the data following: $y_n = \mu_n + \epsilon_n$ with iid gaussian $\epsilon_n$ and a piecewise constant function $\mu_n$, so that we can compare at the end our estimation of $\mu$ with the real $\mu$ defined here.

We start by importing the basic libraries.

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# First, we define our different parameters. 

N = 70  # total number of data
mean = 0  # mean -> used to generate gaussian error
sig2 = 0.3  # var -> used to generate gaussian error


#define the error as an iid Gaussian random variable, N(mean,sig2)
eps = mean + sig2 * np.random.randn(N)
#array of N elmnts

#define a vector x between -10,10 of N elmnts
x = np.linspace(-10,10,N)

#we create a piecewise constant function mu:
def mu(x):
    mu = []
    for i in range(len(x)):
        if x[i] < -3:
            mu.append(-2)
        elif (x[i] >= 0) &  (x[i] < 2) :
            mu.append(6)
        elif x[i] >= 4:
            mu.append(3)
        elif (x[i]>=2) & (x[i] <4):
            mu.append(0)  
        else: 
            mu.append(4)
    return mu

#we generate our data
y = mu(x) + eps

data_x = np.copy(x)
data_y = np.copy(y)
#print(x)
```

```python
# we plot the function mu, and the function y which is mu with noise
plt.plot(x,mu(x), color = 'blue', label = '$\mu$')
plt.plot(x,y, color= 'red', label = 'y')
plt.legend()
```

We can here see the data we generated. The blue function gives us the original function $\mu$, and the red one, the function $\mu$ with noise. 

The aim of this project will be to retrieve the blue function from the red one, using the relaxation algorithm

### 2. Transfoming our problem into Lasso:
In this section, we want to change our problem $$argmin_{\mu} || \textbf{y}-\mu||_2 ^2 + \lambda \sum_{i=1}^{N-1} |\mu_i - \mu_{i+1}|$$ into a multivariate Lasso problem: $$argmin_{\alpha,u} || \textbf{y}-X\alpha||_2 ^2 + \lambda ||\textbf{u}||_1$$ where $\textbf{u}$ is given by the first $N-1$ entries of $\alpha$



The idea here is first to change $ \sum_{i=1}^{N-1} |\mu_i - \mu_{i+1}|$ into $||B\mu||_1$

$B$ is given by a sparse matrix of size $N-1$ x $N$ with $-1$ on the diagonal and $1$ on the upper diagonal; such that $(B\mu)_j = \mu_j - \mu_{j+1}$ for $j = 1,...,n$.
$B$ gives us the finite differences of order one between neighbours. 

```python
# As we saw in class we can go back to a Lasso problem by first defining a matrix B such that(Bf)_i = |fi+1- fi| : the finite differences of order 1 between neighbours.

# We create the B matrix:

M = N-1 # helping for size issues

# We first create a vector (length N-1) of ones. 
v = np.ones(M)

# We then create 2 matrices; B with -1 on the diagonal, C with 1 on the upper diagonal.
B = np.diagflat(-v, k=0) #diag matrix with -1 on the diag (size N-1*N-1)
C = np.diagflat(v, k=1) #diag matrix with 1 on the upper diag (size N*N)

# As we want our final B to be size N-1 x N, we change C into an N-1xN-1 (bcs C is NxN).
D = C[:,0:M] #take only the first M cols of C
D = D[0:M,:] #take only the first M rows of D ->D is size N-1*N-1
K = D+B 
#we now have a N-1 x N-1 matrix as we want, we just need to add the last column.

# Add the last column as a column of zeros.
M,M = K.shape
z = np.zeros((M,1))
# Conccatenate       
F = np.hstack((K,z)) #add last column
# Define last entry as 1.
F[M-1,M] = 1

B = F #created B with size N-1*N
# We now have B.

print(B)
#print(np.shape(B))
```

Our problem can now be written as $argmin_\mu ||\textbf{y}-\mu||_2 ^2 + \lambda||B\mu||_1$



The idea now is to make a change of variable, so that we can write $||\textbf{y}-\mu||_2 ^2$ as $||\textbf{y}-X\alpha||_2 ^2$, so that we can minimize over $\alpha$.

The change of variable is given by 
$\begin{pmatrix} 
	\textbf{u} \\ v
\end{pmatrix} 
= \begin{pmatrix} 
	B\\ (\textbf{0}^T,1)
\end{pmatrix} \mu$

As we want to find $\mu$, we invert our matrix and set $X  = \begin{pmatrix} 
	B\\ (\textbf{0}^T,1)
\end{pmatrix} ^{-1}$

By defining $\alpha = \begin{pmatrix} 
	\textbf{u} \\ v
\end{pmatrix} $, we have $\mu = X\alpha$

```python
#The idea now is to make the change of variable. We first create the matrix X.


# We create the (0.T, 1) vector:
B2 = np.zeros((N,1))
B2[-1] = 1 #last entry = 1
B2 = B2.T # = 1xN

# We concatenate it with B:
XX = np.vstack((B,B2)) # we have created our matrix.
X = np.linalg.inv(XX) # size N*N easy to invert bcause upper diagonal and sparse
# We now have X.


# Now we define u and v:
V = XX.dot(mu(x)) #we compute V, and the first N-1 entries are u, the last entry is v. (V is a vector of size N)
u = V[0:M,]
v = V[-1,]
# We now have X, u and v. by computing X.dot((u,v))we get MU. So we can change in our problem MU by X.dot(u,v)


#we concatenate u and v to have them in one vector called alpha
alpha = np.hstack((u,v)) #vector with N rows, 1 col.
alpha_real = np.copy(alpha)

#define the new problem we want to minimize alpha on:
#PB_multivariate = sum(abs((y - X.dot(alpha)))**2) + lam * sum(abs(alpha[:M]))
# We now have a multivariate problem. We are going to use relaxation to minimize this lasso problem.

# The function PB_multivariate is convex; 
# Furthermore, X has rank N --> the function is strictly convex --> there is uniqueness of the solution. (at most one global minimum)
# It is continue and coercive on R closed -> there is existence of a solution.

print(X)
```

### 3. Solving a univariate Lasso problem

We know that a problem given by $f(\alpha) = ||y-\alpha||_2^2 + \lambda |\alpha|$ has an explicit solution.
The solution is given by $$\alpha_{\lambda,y}^* = sign(y) * \max(|y|- \frac{\lambda}{2},0)$$

So, for a problem given by $f(\alpha) = ||y-x\alpha||_2^2 + \lambda |\alpha|$, where $x$ is a vector in $\mathbb{R}^N $, we have an explicit solution given by : $$\alpha_{\lambda,y}^* = sign(\frac{<y,x>}{<x,x>}) * \max(|\frac{<y,x>}{<x,x>}|- \frac{\lambda}{2<x,x>},0) $$
$$\Leftrightarrow \alpha_{\lambda,y}^* = sign(\frac{<y,x>}{<x,x>}) *  \max(\frac{1}{<x,x>}(|<y,x>|- \frac{\lambda}{2}),0)$$

In this section, we will compute the explicit solution for a univariate Lasso problem.

```python
# We define the explicit solution to the 1D Lasso problem: y,x are the vectors of the Lasso problem, lambda is the penalty.
def oneDLasso_sol(y,x,lam):
    
    alpha_star = np.sign(np.dot(y,x)/np.dot(x,x)) * max(0,np.abs(np.dot(y,x)/np.dot(x,x))-(lam/(2*(np.dot(x,x)))))
    return alpha_star
```

### 4. Transforming a multivariate Lasso problem into N univariate Lasso problems

The aim of Relaxation, is to resolve a multivariate problem by iterating in a cycling way, that is to resolve a minimisation problem for one $\alpha_i$ at a time. 
So, for $i=1,...,N$ we solve the univariate Lasso problem $f(\alpha_i) = ||y-x\alpha_i||_2^2 + \lambda |\alpha_i|$ with the oneDLasso_sol function created.
We then get a vector with all the best alphas we could get.

The idea of relaxation, is that we do this several times. That is, we run our relaxation problem for $i = 1,....,N,1,....N,1,....N,$ etc.


```python
# RELAXATION: at every iteration, we want to resolve a univariate problem.
# We change the equation so that we only have one entry of alpha to minimize onto.


# DO WE HAVE UNIQUENESS AND EXISTENCE FOR THIS PROBLEM? Yes because we do for the multivariate problem: same reasons.
# IS THERE A SOLUTION IN EXACT FORM? Yes: given by the oneDLasso_sol.

def relaxation(y, X, alpha, lam):
    alpha_starr = []  # empty vector in which we will append all the solutions for each univariate Lasso problem.
    
    #we create our univariate problem and resolve it with the function oneDLasso_sol
    #we append the best alphas for each entry of alpha in the vector alpha_starr.
    for i in range(len(alpha)):
        if i < N-1:
            X_i = np.delete(X, i, 1)  # X without the ith column
            alpha_i = np.delete(alpha, i)  # alpha without the ith entry
            XA_i = X_i.dot(alpha_i)  # comuptation of X_i * alpha_i
            y_new = y - XA_i # the new y for the univariate problem
            x_new = X[:, i]  # calling the ith column; the new x for the univariate problem.
            sol_oneD = oneDLasso_sol(y_new, x_new, lam)  # we use the explicit solution given for a univariate Lasso problem, on the new y and x.
            alpha_starr.append(sol_oneD)
            alpha[i]= sol_oneD #we use this solution as the new ith entry for the initial point in the next loop.
        else : 
            X_n = np.delete(X, -1, 1) #as before but for the last column
            alpha_n = np.delete(alpha, -1)
            XA_n = X_n.dot(alpha_n)
            y_new = y - XA_n
            x_new = X[:, -1]
            
            # Least square solution for the last entry of alpha. (because not a Lasso prob anymore but a Least-Sq)
            A = ((x_new.T).dot(x_new))**(-1)
            A2 = (x_new.T).dot(y_new)
            ls_sol = A*(A2)
            alpha[i]= ls_sol
            alpha_starr.append(ls_sol)
    return alpha_starr


loop = range(30) # number of times we want our relaxation to loop

#we here define a function that will loop the relaxation the amount of times we want. 
#for instance, if we want our relaxation algorithm to run 3 times : i = 1,...,N,1,...,N,1,....,N. we define loop as range(3).
def looprelax(y,X,alpha, lam):
    alpha_dep = [alpha]
    for k in loop: #looping times on our relaxation
        j = relaxation(y,X,alpha_dep[k],lam)
        alpha_dep.append(j)
    return alpha_dep[-1]

#NB each time we loop the relaxation, we use the result of the last as the new initial point for the next relaxation.
        
```

### 5. Plotting a solution:

From the $\alpha$ we found from the relaxation algorithm, we want now to retrieve our $\mu$ function. We create a function, that given our data ($y$, $X$, $\alpha$, $\lambda$) gives us an apporximation of $\mu$

```python
def solution(y,X,lam, initial_alpha):
    alpha_star = looprelax(y, X,initial_alpha,lam)
    mu_star = X.dot(alpha_star)
    return mu_star
```

The idea now is to plot our solution, and to try it for different $\lambda$ and initial vectors $\alpha$.
This can make us understand the importance of the selection of the initial point.

```python
# We plot the function for several lambdas, to see the difference. 
lam = np.linspace(0,20,50)

print('the lambda where the alpha vector is zero is given by %s' %np.linalg.norm((X.T).dot(y),np.inf))

# we here use as an initial point, the true alpha, given by the computation of mu. 
initial_alpha = alpha_real
for l in lam:
    plt.figure()
    plt.plot(x,solution(y,X,l,initial_alpha), color = 'red', label = 'mu_star for lambda =%s '%l)
    plt.plot(x,mu(x), color = 'blue', label = '$\mu$')
    plt.legend()

# we also plot here the solution with the penalty that is equal to the norm of X.T * y.   
# we know that passed this value of lambda, all entries of alpha are 0, and so the approximation of the problem is not good.
# we can see it in the plot, that compared to the others, the approximation of mu is really bad.
plt.figure()
plt.plot(x,mu(x), color = 'blue', label = '$\mu$')
plt.plot(x,solution(y,X,np.linalg.norm((X.T).dot(y-np.mean(y)),np.inf),initial_alpha),color ='green', label = 'mu_star for lambda =%s '%np.linalg.norm((X.T).dot(y),np.inf))
plt.legend()
```

$\textbf{Interpretation:}$ 

We see in these graphs the approximation of $\mu$ for several $\lambda$, that we got from the $\alpha$ estimated by the relaxation. Let's point out two main observations: 

1. We first notice that for different initial vectors $\alpha$, the approximated function reacts differently:

We see that, sometimes, the approximation is really really bad. Especially when $\alpha$ is chosen randomly.
Take for example $\alpha$ = the vector of ones. 
Take as a comparison $\alpha$ = alpha_real (the "real" $\alpha$  computed initially, to extract the $X$ matrix for the problem). We immediately see that for the "real" $\alpha$ the approximation is much better. That is because, it is much closer to the solution we tried to obtain.

From an initial point really far from our goal, it is hard to achieve a good approximation of it.
This makes us understand the importance of choosing wisely our initial point.


2. We observe that for small $\lambda$, the approximation is good, but not that good. The largest $\lambda$ gets, the better the approximation fits our $\mu$ function.
The problem is that in the last graph, we see that for $\lambda$ $\textbf{too}$ big, there is literally no approximation at all.

We can see there is a correlation between the approximation, the $\lambda$ and the initial point $\alpha$.
So the following question is: how to know from a given $\lambda$, which initial point to choose?

### 6. Warm start:

In this section, we want to use the warm start algorithm, to find out the initial point that fits the best according to the given $\lambda$.

To do the warm start, we begin with the $\lambda$ such that all entries of $\alpha$ are zero. We know that such a $\lambda$ exists, because, at one point, the problem is so penalized, that all entries of $\alpha$ are negligeable. This $\lambda$ is given by $\lambda = max|X^Ty|$. 

We first run our algorithm with this $\lambda$, and then decrease it slowly until it is close to zero. We do not look pass this $\lambda$, because we know that all entries of $\alpha$ will be zero from this point on, and so our approximation will be useless.
We use as first initial point the vector $\alpha$ given by the $0$ vector, as we know that for that $\lambda$, it is a really good initial value to begin the relaxation with. 

The idea then is to run the relaxation each time while decreasing the $\lambda$ and updating the initial values for $\alpha$, so that they are closer to the solution. 

```python
lamrange = range(50) #number of lambdas we want to observe.

def warmstart(y, X):
    lam = np.linalg.norm((X.T).dot(y), np.inf) #lam_0
    alpha_init = np.zeros(N).T #initializing with a vector of zeros.
    alpha_best = [alpha_init] #appending matrix's 0_entry
    LAM = [lam] #same
    for i in lamrange: #amount of lambda does not depend on how large the data set is. 
        k = looprelax(y,X, alpha_best[i],lam)
        alpha_best.append(k)
        new_lam = 9/10 * lam
        kk = looprelax(y, X, alpha_best[i], new_lam)
        alpha_best[i+1] = kk
        lam = new_lam
        
        if lam != LAM[-1]:
            LAM.append(lam)
        #print(" lambda = %s " %LAM[i] )
        #print(" gives us de value of the alphas: %s " %alpha_best[i])

    return [alpha_best,LAM]

[alpha_best,lam_best] = warmstart(y, X)

#return a vector with best alphas according to lambdas.
```

### 7. Comparing with other algorithms and plotting

Here we want to define the running mean algorithm. With it we can estimate the $\mu$ function at any point $x$.
The running mean estimator is given by: $$\hat{\mu}_{\eta}^{RM}(x) = \frac{\sum_{j=1}^n y_j 1_{|x_j-x|\leq\eta/2}}{\sum_{j=1}^n 1_{|x_j-x|\leq\eta/2}}$$

To estimate $\mu(x)$ at a point $x$ outside the range of the data $[x_{(1)},x_{(N)}]$ we define: $$ \hat{\mu}_{\eta}^{RM}(x) = \left\{
    \begin{array}{ll}
        \hat{\mu}_{\eta}^{RM}(x_{(1)})  \hspace{1cm} x \leq x_{(1)}\\
        \hat{\mu}_{\eta}^{RM}(x_{(N)}) \hspace{1cm} x \geq x_{(N)}
    \end{array}
\right. $$

```python
def running_mean(eta,x,vec_x,vec_y):
    
    N = len(vec_x) #N = nb of data.
    
    a = 0 
    b = 0 
    
    #if x >= x_N: we compute rm(x_N). So we define x as being the min between its value and the vect.
    x = min([x,vec_x[-1]])
    #if x <= x_1: rm(x_1). So x is its value or x_1 if it is smaller.
    x = max([x,vec_x[0]])

    
    for j in range(N):

        if abs(vec_x[j] - x) <= eta/2: #if 1 function is 1.
            a = vec_y[j] + a  #sum above
            b = 1 + b #sum below
            
   #compute rm
    if b != 0:    
        running_mean = a / b
        
    else: #if b = 0, all points around x are further that a lam/2 distance. 
        # if it is the case, we choose the x the closest = that has the smallest distance.
        min_dist = np.argmin(abs(vec_x-x))
        running_mean = vec_y[min_dist]  
        
    return running_mean


#we put all our estimations in a vector stock_rm to be able to plot it.

def stock_rm(eta,x,vec_x,vec_y):
    stock = []
    for i in range(len(x)):
        stock.append(running_mean(eta,x[i],vec_x,vec_y))
        
    return stock      
```

```python
#Plotting the RM: 

eta1 = 0.5
eta2 = 6
x = data_x
y = data_y
#x's we want to estimate the mu function on.
plot = np.linspace(min(x[[0,-1]])*0.8,max(x[[0,-1]])*1.2, num = N)

plt.figure('Estimation by running mean')
#plt.plot(x,y,'o',label = 'data')
plt.plot(x,mu(x),label='function mu')
plt.plot(plot, stock_rm(eta1,plot,x,y), label = 'running mean $\eta$ =0.1')
plt.plot(plot, stock_rm(eta2,plot,x,y), label = 'running mean $\eta$ = 6')

plt.legend()

```

We notice here that the running mean algorithm gives us a good approximation for a small window width $\eta$.

By plotting the approximation for two different $\eta$ we can see that for bigger values of $\eta$, the running mean doesn't do a good job for the approximation. It is quite logical, because the more the window is big, the less accurate the average of the values will be; because there will be more values that will be far from each other, than in a small window.

```python
#Plotting the Relaxation:

[alpha_best,lam_best] = warmstart(y, X)

x = data_x
y = data_y

for i in lamrange:
    lam = lam_best[i]
    f = alpha_best[i]
    relax = solution(y,X,lam,f)
    eta = 0.1


    plt.figure('Estimation by relaxation')
    #plt.plot(x,y,color = 'red',label='data')
    #plt.plot(x,y,'o',label = 'data')
    plt.plot(x,mu(x),label='function $\mu$')
    plt.plot(x, relax, color = 'blue', label = '$\mu^*$ _relaxation') #plotting the relaxation for each lambda of the loop
    plt.plot(plot, stock_rm(eta,plot,x,y), label = '$\hat{\mu}$ _running mean') #plotting the running mean for each lambda of the loop
    plt.title('lam = %s' %lam)

    plt.legend()
    plt.show()
```

We here plotted:
- the initial function $\mu$
- the function $\mu^*$ computed with the relaxation, plotted for different $\lambda$ for which we chose the matching initial points $\alpha$ obtained with the warm start.
- the function $\hat{\mu}$ computed with the running mean.

We here see that the running mean is much better for $\eta$ small, as it represents the width of a window on which we compute the average of the values in. Of course, the more the window is big, the less accurate the approximation will be. Moreover, if $\eta$ is too small, the approximation is still noisy...

On the other hand, the approximation given by the relaxation algorithm gets better at a small $\lambda$ but not $\textbf{too}$ small. This is normal, as we don't want to solve a Least-Square problem (ie. when $\lambda$ is zero) neither we want our problem to be too penalized (ie. when $\lambda$ is huge).


We here plot the function $\mu = \cos(x) * x^2$ and look at the difference of the approximation given by the running mean and the relaxation:

```python
x = data_x

def mu(x):
    k = np.cos(x)* x**2
    return k
y = mu(x) + eps

#Plotting the Relaxation:

[alpha_best,lam_best] = warmstart(y, X)



for i in lamrange:
    lam = lam_best[i]
    f = alpha_best[i]
    relax = solution(y,X,lam,f)
    eta = 0.5

    plt.figure('Estimation by relaxation')
    #plt.plot(x,y,color = 'red',label='data')
    #plt.plot(x,y,'o',label = 'data')
    plt.plot(x,mu(x),label='function $\mu$')
    plt.plot(x, relax, color = 'blue', label = '$\mu^*$ _relaxation') #plotting the relaxation for each lambda of the loop
    plt.plot(plot, stock_rm(eta,plot,x,y), label = '$\hat{\mu}$ _running mean') #plotting the running mean for each lambda of the loop
    plt.title('lam = %s' %lam)

    plt.legend()
    plt.show()

```

We can see here that the running mean is much better than for a piecewise constant function. However, the relaxation still does a good job when the penalty is small.
As the running mean is a much faster algorithm, here we would prefer it. 

### Conclusion
To sum up, we can see that the relaxation algorithm, with initial points defined by the warm start, gives us a good approximation of our problem. It does even a better one than the running mean.
However, in some cases, such as when $\mu$ is not piecewise constant, we would prefer the running mean estimator to find our $\mu$ as it is efficient and less expensive to compute.
